```python
AED.groupby('場所分類').size()
```

```python
from shapely.geometry import Point
gdf = gpd.GeoDataFrame(
    AEDD.drop(['場所LAT','場所LNG'], axis=1),
    crs={'init': 'epsg:4326'},
    geometry=[Point(xy) for xy in zip(AED['場所LAT'],AED['場所LNG'])])
```

```python
A1=AEDD.loc[AEDD['其他類型'] == '醫院']
A2=AEDD.loc[AEDD['其他類型'] == '醫療']
A3=AEDD.loc[AEDD['其他類型'] == '衛生所']

```

```python
AEDD=pd.concat([A1,A2,A3],axis=0).loc[:,['場所名稱','場所LAT','場所LNG']]
from shapely.geometry import Point
gdf = gpd.GeoDataFrame(
    AEDD.drop(['場所LAT','場所LNG'], axis=1),
    crs={'init': 'epsg:4326'},
    geometry=[Point(xy) for xy in zip(AEDD['場所LAT'],AEDD['場所LNG'])])
```
